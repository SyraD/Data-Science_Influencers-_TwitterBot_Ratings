## How Many of the Top 123 Data Science Twitter Accounts are Bots?

Data science is trending. 

Each time that I posted a Tweet containing phrases like "big data" or "analysis", a flood of twitter followers, list adds, and "likes" would appear at suspiciously super-human speeds. This left me wondering, how much of the Twitter activity surrounding data science is automated? 

Is it possible that the data science social network has become a set of algorithm-based TwitterBots interacting with other automated TwitterBots? 

The same accounts that quickly followed and unfollowed me based on keywords from my tweets were spreading headlines about the top 10 or top 120 most influential data scientists based on twitter rankings. 

"Curiouser and curiouser", I thought.

This notebook consists of a quick & dirty pet project to answer:
How many of the top ranked influencer data science twitter accounts were automated bots? 

TwitterBot ranking from  http://truthy.indiana.edu/botornot/ on Twitter @TruthyBotOrNot
Influencer data from: http://most-influential-data-science-accounts-on.silk.co/

```python
import pandas as pd
import botornot
import tweepy
import time
import random
```

```python
twitter_app_auth = {
    'consumer_key': 'XXX',
    'consumer_secret': 'XXX',
    'access_token': 'XXX
    'access_token_secret': 'XXX',
  }
bon = botornot.BotOrNot(**twitter_app_auth)    
```

```python
full_output = []
```

```python
for i in range(len(influencers)):
    try:
        output_json = bon.check_account(influencers[i])
        full_output.append(output_json)
        print influencers[i], "done."
        sleep_time = random.randint(5,15) #fixed apparent jamming issue
        print "sleeping for",  sleep_time
        time.sleep(sleep_time)
            
    except:
        print " ERROR: ",  influencers[i], "*********************"
        
# Errors in 2 accounts that no longer exist or are modified timefulapp & pivotaldatalabs
```

```python
full_output
```

```python
df = pd.DataFrame(columns=('screen_name', 'user_id', 'score',
                               'content_classification','friend_classification',
                               'sentiment_classification','temporal_classification',
                               'user_classification'))
```

```python
def into_pandas(user_output, row_num, df):
    df.loc[row_num,'content_classification'] = user_output[u'categories'][u'content_classification']
    df.loc[row_num,'friend_classification'] = user_output[u'categories'][u'friend_classification']
    df.loc[row_num,'network_classification'] = user_output[u'categories'][u'network_classification']
    df.loc[row_num,'sentiment_classification'] = user_output[u'categories'][u'sentiment_classification']
    df.loc[row_num,'temporal_classification'] = user_output[u'categories'][u'temporal_classification']
    df.loc[row_num,'user_classification'] = user_output[u'categories'][u'user_classification']
    df.loc[row_num,'screen_name'] = user_output[u'meta'][u'screen_name']
    df.loc[row_num,'user_id'] = user_output[u'meta'][u'user_id']
    df.loc[row_num,'score'] = user_output[u'score']
```

```python
for i in range(len(full_output)):
    into_pandas(full_output[i], i, df)
```
